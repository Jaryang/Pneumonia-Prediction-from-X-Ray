```python
import matplotlib.pyplot as plt
import pprint as pp
import json
```

# Trials on Dropout Rates

```python
with open('models/hyperparameters/dropout/result_0.6do.txt', 'r') as f:
    dropout_six = json.loads(f.read())
    print(dropout_six.keys())

with open('models/hyperparameters/dropout/result_0.7do.txt', 'r') as f:
    dropout_seven = json.loads(f.read())
    print(dropout_seven.keys())

with open('models/hyperparameters/dropout/result_0.8do.txt', 'r') as f:
    dropout_eight = json.loads(f.read())
    print(dropout_eight.keys())
```

```python
# Plot Accuracy Over Epochs
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
epochs_array = [i for i in range(1, 26)]

# train: solid
ax.plot(epochs_array, dropout_six['train_accuracies'], label = '0.6_train_acc', linestyle = 'dashed')
ax.plot(epochs_array, dropout_seven['train_accuracies'], label = '0.7_train_acc', linestyle = 'dashed')
ax.plot(epochs_array, dropout_eight['train_accuracies'], label = '0.8_train_acc', linestyle = 'dashed')
# validation: dashed
ax.plot(epochs_array, dropout_six['val_accuracies'], label = '0.6_val_acc', linestyle = 'solid')
ax.plot(epochs_array, dropout_seven['val_accuracies'], label = '0.7_val_acc', linestyle = 'solid')
ax.plot(epochs_array, dropout_eight['val_accuracies'], label = '0.8_val_acc', linestyle = 'solid')


ax.set_title('Accuracies of Different Dropout Rates', fontsize=30)
ax.legend(loc="lower right", fontsize=16)
ax.set_xlabel("epochs", fontsize=16)
ax.set_ylabel("accuracy", fontsize=16)
plt.show()
```

```python
# Plot Accuracy Over Epochs
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
epochs_array = [i for i in range(1, 26)]

# train: solid
ax.plot(epochs_array, dropout_six['train_losses'], label = '0.6_train_loss', linestyle = 'dashed')
ax.plot(epochs_array, dropout_seven['train_losses'], label = '0.7_train_loss', linestyle = 'dashed')
ax.plot(epochs_array, dropout_eight['train_losses'], label = '0.8_train_loss', linestyle = 'dashed')
# validation: dashed
ax.plot(epochs_array, dropout_six['val_accuracies'], label = '0.6_val_loss', linestyle = 'solid')
ax.plot(epochs_array, dropout_seven['val_accuracies'], label = '0.7_val_loss', linestyle = 'solid')
ax.plot(epochs_array, dropout_eight['val_accuracies'], label = '0.8_val_loss', linestyle = 'solid')


ax.set_title('Losses of Different Dropout Rates', fontsize=30)
ax.legend(loc="lower right", fontsize=16)
ax.set_xlabel("epochs", fontsize=16)
ax.set_ylabel("loss", fontsize=16)
plt.show()
```

# Trials on Epochs

```python
with open('models/hyperparameters/epochs/result_five_epoch.txt', 'r') as f:
    epoch_five = json.loads(f.read())
    print(epoch_five.keys())

with open('models/hyperparameters/epochs/result_ten_epoch.txt', 'r') as f:
    epoch_ten = json.loads(f.read())

with open('models/hyperparameters/epochs/result_fifteen_epoch.txt', 'r') as f:
    epoch_fifteen = json.loads(f.read())

with open('models/hyperparameters/epochs/result_twentyfive_epoch.txt', 'r') as f:
    epoch_twentyfive = json.loads(f.read())
```

```python
x_5 = [i for i in range(1, 6)]
x_10 = [i for i in range(1, 11)]
x_15 = [i for i in range(1, 16)]
x_25 = [i for i in range(1, 26)]
```

```python
# Plot Accuracy Over Epochs
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))

# train: solid
ax.plot(x_5, epoch_five['train_accuracies'], label = '5_train_acc', linestyle = 'dashed')
ax.plot(x_10, epoch_ten['train_accuracies'], label = '10_train_acc', linestyle = 'dashed')
ax.plot(x_15, epoch_fifteen['train_accuracies'], label = '15_train_acc', linestyle = 'dashed')
ax.plot(x_25, epoch_twentyfive['train_accuracies'], label = '25_train_acc', linestyle = 'dashed')
# validation: dashed
ax.plot(x_5, epoch_five['val_accuracies'], label = '5_val_acc', linestyle = 'solid')
ax.plot(x_10, epoch_ten['val_accuracies'], label = '10_val_acc', linestyle = 'solid')
ax.plot(x_15, epoch_fifteen['val_accuracies'], label = '15_val_acc', linestyle = 'solid')
ax.plot(x_25, epoch_twentyfive['val_accuracies'], label = '25_val_acc', linestyle = 'solid')


ax.set_title('Accuracies of Different Epochs', fontsize=30)
ax.legend(loc="lower right", fontsize=16)
ax.set_xlabel("epochs", fontsize=16)
ax.set_ylabel("accuracy", fontsize=16)
plt.show()
```

```python
# Loss Over Epochs
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))

# train: solid
ax.plot(x_5, epoch_five['train_losses'], label = '5_train_loss', linestyle = 'dashed')
ax.plot(x_10, epoch_ten['train_losses'], label = '10_train_loss', linestyle = 'dashed')
ax.plot(x_15, epoch_fifteen['train_losses'], label = '15_train_loss', linestyle = 'dashed')
ax.plot(x_25, epoch_twentyfive['train_losses'], label = '25_train_loss', linestyle = 'dashed')
# validation: dashed
ax.plot(x_5, epoch_five['val_losses'], label = '5_val_loss', linestyle = 'solid')
ax.plot(x_10, epoch_ten['val_losses'], label = '10_val_loss', linestyle = 'solid')
ax.plot(x_15, epoch_fifteen['val_losses'], label = '15_val_loss', linestyle = 'solid')
ax.plot(x_25, epoch_twentyfive['val_losses'], label = '25_val_loss', linestyle = 'solid')


ax.set_title('Losses of Different Epochs', fontsize=30)
ax.legend(loc="lower right", fontsize=16)
ax.set_xlabel("epochs", fontsize=16)
ax.set_ylabel("loss", fontsize=16)
plt.show()
```
